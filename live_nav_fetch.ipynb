Fetch NAV for 5 key schemes: SBI Bluechip (119551), ICICI Bluechip (120503), Nippon Large Cap (118632), Axis Bluechip (119092), Kotak Bluechip (120841).


```python
import requests
import pandas as pd
import os

# Save data/raw folder 
os.makedirs("data/raw", exist_ok=True)

# Scheme names aur unke AMFI codes
schemes = {
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_Large_Cap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841
}

#  API call for every Schemes
for scheme_name, code in schemes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    if response.status_code == 200:

        data = response.json()

        # COnvert NAV History into Data frame
        df = pd.DataFrame(data["data"])

        # Save the Data Frame 
        file_path = f"data/raw/{scheme_name}.csv"
        df.to_csv(file_path, index=False)

        print(f"{scheme_name} data saved successfully.")

    else:
        print(f"Failed to fetch {scheme_name}")
```
